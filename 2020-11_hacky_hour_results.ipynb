# Hacky hour survey results, 2020-11-07

```python
from textwrap import wrap
import pandas as pd
import hvplot.pandas
```

## Load data

```python
df = pd.read_csv("results-survey457853.csv", delimiter=";")
display(df)
display(df.columns)
```

## General comments

```python
print("\n\n".join(
    map(
        lambda s: "\n".join(wrap(s, 70)),
        df["Feel free to leave any comment here."].dropna()
    )
))
```

## Roles

```python
display(
    df["What\'s your role?"].dropna().value_counts().sort_values().hvplot.barh()
)
comments = ';  '.join(df["What's your role? [Comment]"].dropna())
print(f"\nOthers:  {comments}")
```

## Interest

```python
interest_matrix = df.filter(like="What application and language")
interest_matrix = interest_matrix.rename(columns=lambda s: tuple(s.split("?")[-1][2:-1].split("][")[::-1]))
interest_matrix = interest_matrix.sum().to_frame()
interest_matrix = interest_matrix.set_index(pd.MultiIndex.from_tuples(interest_matrix.index))
interest_matrix = interest_matrix.unstack(0)[0].astype("int")

display(interest_matrix)
```

```python
display(interest_matrix.hvplot.heatmap(cmap="viridis"))
```

## Interest: Comments

```python
print("\n\n".join(
    map(
        lambda s: "\n".join(wrap(s, 70)),
        df["If you choose other, please elaborate."].dropna()
    )
))
```

## Mode of attendance

```python
how_often = df["How often would you likely attend?"].dropna().value_counts().sort_values()
display(how_often)
display(how_often.hvplot.barh())
```

```python
give_receive = df.filter(like="Would you be ").dropna().replace({"Yes": True, "No": False})
give_receive = give_receive.rename(columns=lambda s: s.split("?")[-1][2:-1])
give_receive = give_receive.sum()
display(give_receive)
display(give_receive.hvplot.barh().opts(title="How would you participate?"))
```

## Best time slots

```python
best_times = df.filter(like="When could you best attend")
best_times = best_times.rename(columns=lambda s: tuple(s.split("?")[-1][2:-1].split("][")[::-1]))
best_times = best_times.sum().to_frame()
best_times = best_times.set_index(pd.MultiIndex.from_tuples(best_times.index))
best_times = best_times.unstack(0)[0][["Mon", "Tue", "Wed", "Thu", "Fri"]].astype("int")

display(best_times)
```

```python
display(best_times.hvplot.heatmap(cmap="viridis"))
```

## Best time slots (students)

This is important as we want to make it especially easy for students to come to the hacky hour.

```python
student_filter = df["What\'s your role?"].fillna("").apply(lambda s: "Student" in s)
student_filter
```

```python
best_times_students = df.where(student_filter).filter(like="When could you best attend")
best_times_students = best_times_students.rename(columns=lambda s: tuple(s.split("?")[-1][2:-1].split("][")[::-1]))
best_times_students = best_times_students.sum().to_frame()
best_times_students = best_times_students.set_index(pd.MultiIndex.from_tuples(best_times_students.index))
best_times_students = best_times_students.unstack(0)[0][["Mon", "Tue", "Wed", "Thu", "Fri"]].astype("int")

display(best_times_students)
```

```python
display(best_times_students.hvplot.heatmap(cmap="viridis"))
```
